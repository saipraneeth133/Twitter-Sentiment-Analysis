# Preprocessing

for Preprocessing of tweets, following methods are considered



*   Lemitization
*   Removal of Stop Words
*   Remval of Punctuation
*   Removing Hashtags from the words (Remove hashtag and seperate word)
*   Convertion of Emoji's to text
*   Spell correction of words







```python

def preprocess(doc):
    emoji_chars = emoji.EMOJI_ALIAS_UNICODE.values()
    tag_map = defaultdict(lambda: wn.NOUN)
    tag_map['J'] = wn.ADJ
    tag_map['V'] = wn.VERB
    tag_map['R'] = wn.ADV
    spell = Speller(lang='en')
    final1 = []
    stopwords_list=(stopwords.words('english'))
    alphabets = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't',
                 'u', 'v', 'w', 'x', 'y', 'z', '<START>','<NULL>','<END>','stop' , '"', '<user>']
    punch=list(string.punctuation)
    stopwords_list.extend(alphabets)
    stopwords_list.extend(punch)
    for index, entry in enumerate(doc):
        final_words = []
        word_lemmatized = WordNetLemmatizer()
        for word, tag in pos_tag(entry):
            if word not in stopwords_list:
                if word not in alphabets:
                    word_final = word_lemmatized.lemmatize(word, tag_map[tag[0]])
                    if word_final in emoji_chars:
                      word_final=str(unicodedata.name(word_final)).lower()
                    else:
                      pass

                    tweet = ''.join(re.sub('[^A-Za-z0-9]+', '', word_final))
                    final_words.append(spell(tweet))
        final1.append(final_words)
        
    return final1

```

# Feature Extraction

the product of Term Frequency and Inverse Doucument frequency will give the measure of how frequent the word is in a document multiplied by how unique the word is, giving rise to Term Frequency-Inverse Document Frequency(TF-IDF) measure.


*   Defining Function for Feature Extraction (Dimensionality Reduction/Pattern matching)
*   TfidfVectorizer is used to perform the activity



```python
def vector(data, fulldata):
  vectorizer = TfidfVectorizer (max_features=10000, ngram_range=(1,2), min_df=7, max_df=0.08, stop_words=stopwords.words('english'))
  vectorizer.fit(fulldata)
  vector = vectorizer.transform(data)
  vector_array=vector.toarray()
  return(vector_array)
```


*   Function For reviewing generated Vocabulary
*   Prediction accuracy can be improved by properly building the vocubulary



```python
def vector_word(data, fulldata):
  vectorizer = TfidfVectorizer(max_features=10000,ngram_range=(1,1), smooth_idf=False , min_df=0.00001, norm='l2', max_df=0.1, binary=True, sublinear_tf=True)
  vectorizer.fit(fulldata)
  vector = vectorizer.transform(data)
  vectorizer.vocabulary_
  vector_array=vector.toarray()
  return(vectorizer.vocabulary_)
```
